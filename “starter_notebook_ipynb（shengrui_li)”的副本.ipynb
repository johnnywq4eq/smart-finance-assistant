### Foundation Data Processing Functions

````python
# 🤖 AI Collaboration: Data Loading and Cleaning Function
# Ask AI to help you create a robust data loading function

def load_and_clean_transaction_data(file_path):
    try:
        # Load data
        df = pd.read_csv(file_path)

        # Clean Amount column
        df['Amount'] = df['Amount'].replace('[\$,]', '', regex=True)
        df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')

        # Remove invalid rows
        df = df.dropna(subset=['Amount'])

        return df

    except Exception as e:
        print("Error:", e)
        return None


# Test your function
df_sample.to_csv("sample_transactions.csv", index=False)
clean_data = load_and_clean_transaction_data("sample_transactions.csv")
print(clean_data)
````

````python
# 🤖 AI Collaboration: Spending Analysis Function
# Ask AI to help you create comprehensive spending analysis

def analyze_spending_patterns(df):
    """
    Analyze spending patterns and generate business insights.

    Args:
        df: Cleaned transaction DataFrame

    Returns:
        dict: Analysis results and insights
    """
    # Keep only positive spending for category analysis
    positive_spending = df[df['Amount'] > 0]

    # Total spending
    total_spending = positive_spending['Amount'].sum()

    # Spending by category
    category_summary = positive_spending.groupby('Category')['Amount'].sum().sort_values(ascending=False)

    # Percentage by category
    category_percentages = (category_summary / total_spending * 100).round(2)

    # Highest spending category
    top_category = category_summary.idxmax()
    top_amount = category_summary.max()

    # Average transaction
    average_transaction = positive_spending['Amount'].mean().round(2)

    # Simple business insight
    insight = f"The highest spending category is {top_category} with ${top_amount:.2f}. This suggests the user may review spending in this area."

    return {
        "total_spending": round(total_spending, 2),
        "category_summary": category_summary,
        "category_percentages": category_percentages,
        "top_category": top_category,
        "top_amount": round(top_amount, 2),
        "average_transaction": average_transaction,
        "insight": insight
    }

# Test your function
analysis = analyze_spending_patterns(clean_data)
print(analysis)
# analysis = analyze_spending_patterns(clean_data)
# print(analysis)analysis = analyze_spending_patterns(clean_data)

print("Total spending:", analysis["total_spending"])
print("\nCategory summary:")
print(analysis["category_summary"])
print("\nCategory percentages:")
print(analysis["category_percentages"])
print("\nTop category:", analysis["top_category"])
print("Top amount:", analysis["top_amount"])
print("Average transaction:", analysis["average_transaction"])
print("\nInsight:", analysis["insight"])
````

````python
# 🤖 AI Collaboration: Business Insights Generator
# Ask AI to help create professional financial recommendations

def generate_financial_recommendations(analysis_data):
    """
    Generate actionable financial recommendations based on spending analysis.

    Args:
        analysis_data: Dictionary with spending analysis results

    Returns:
        str: Formatted recommendations report
    """
    top_category = analysis_data["top_category"]
    top_amount = analysis_data["top_amount"]
    total_spending = analysis_data["total_spending"]
    average_transaction = analysis_data["average_transaction"]

    recommendations = []

    recommendations.append(f"Total spending is ${total_spending:.2f}.")
    recommendations.append(f"The highest spending category is {top_category} with ${top_amount:.2f}.")
    recommendations.append(f"The average transaction amount is ${average_transaction:.2f}.")

    if top_category == "Entertainment":
        recommendations.append("You may reduce entertainment spending by setting a monthly limit and focusing on essential expenses first.")
    elif top_category == "Groceries":
        recommendations.append("You may save money on groceries by planning meals, comparing prices, and avoiding unnecessary purchases.")
    elif top_category == "Transport":
        recommendations.append("You may reduce transport spending by using cheaper travel options or planning trips more efficiently.")
    elif top_category == "Coffee":
        recommendations.append("Frequent coffee purchases may seem small, but they can add up over time. Making coffee at home may help reduce costs.")
    else:
        recommendations.append("Review this category carefully and consider whether some of the spending can be reduced.")

    if total_spending > 100:
        recommendations.append("Overall spending is relatively high in this sample, so creating a simple weekly budget may help improve financial control.")
    else:
        recommendations.append("Overall spending is at a moderate level, but regular tracking is still important for long-term savings.")

    return "\n".join(recommendations)

# Test your function
recommendations = generate_financial_recommendations(analysis)
print(recommendations)
# recommendations = generate_financial_recommendations(analysis)
# print(recommendations)
````

## Custom Financial Tools

````python
def create_savings_calculator_tool(current_savings, monthly_saving, target_amount):
    """
    Calculate how long it takes to reach a savings goal.

    Args:
        current_savings (float): Current amount saved
        monthly_saving (float): Amount saved per month
        target_amount (float): Target savings goal

    Returns:
        str: Result message
    """

    try:
        # Input validation
        if monthly_saving <= 0:
            return "Monthly saving must be greater than 0."

        if target_amount <= current_savings:
            return "You have already reached your savings goal!"

        # Calculate months needed
        remaining_amount = target_amount - current_savings
        months_needed = remaining_amount / monthly_saving

        # Format output
        return f"You need approximately {months_needed:.1f} months to reach your savings goal of ${target_amount:.2f}."

    except Exception as e:
        return f"Error: {e}"
print(create_savings_calculator_tool(500, 100, 2000))
````
